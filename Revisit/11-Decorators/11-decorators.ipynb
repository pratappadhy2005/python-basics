```python
def div(a, b):
    return a / b

def smart_div(func):
    def inner(a, b):
        if a < b:
            a, b = b, a
        return func(a, b)
    return inner

newdiv = smart_div(div)
newdiv(2,4)
```

```python
import time
from datetime import datetime, timedelta

def timer_dec(base_func):
    def enhanced_func(*args, **kwargs):
        start_time = time.time()
        result = base_func(*args, **kwargs)
        end_time = time.time()
        print(f"Time taken: {end_time - start_time} seconds")
        return result
    return enhanced_func

@timer_dec
def brew_tea(tea_type, steep_time):
    print(f"Brewing {tea_type}...")
    time.sleep(steep_time)
    print(f"{tea_type} is ready!")

@timer_dec
def brew_coffee(coffee_type, steep_time):
    print(f"Brewing {coffee_type}...")
    time.sleep(steep_time)
    print(f"{coffee_type} is ready!")

@timer_dec
def brew_matcha(matcha_type, steep_time):
    print(f"Brewing {matcha_type}...")
    time.sleep(steep_time)
    print(f"{matcha_type} is ready!")
    return f"Drink matcha {matcha_type} by {datetime.now() + timedelta(seconds=steep_time)}"

@timer_dec
def brew_plain_tea():
    print("Brewing plain tea...")
    time.sleep(3)
    print("Tea is ready!")

brew_tea("Green Tea", 3)
brew_coffee("Espresso", 3)
print(brew_matcha("Matcha", 3))
brew_plain_tea()
```
