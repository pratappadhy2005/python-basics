```python
def div(a, b):
    return a / b

def smart_div(func):
    def inner(a, b):
        if a < b:
            a, b = b, a
        return func(a, b)
    return inner

newdiv = smart_div(div)
newdiv(2,4)
```

```python
import time

def timer_dec(base_func):
    def enhanced_func():
        start_time = time.time()
        base_func()
        end_time = time.time()
        print(f"Time taken: {end_time - start_time} seconds")
    return enhanced_func

@timer_dec
def brew_tea():
    print("Brewing tea...")
    time.sleep(3)
    print("Tea is ready!")


brew_tea()
```
