```python
from abc import ABC, abstractmethod

## Abstract class
class Vehicle(ABC):

    def drive(self):
        print("Driving the vehicle")

    @abstractmethod
    def start_engine(self):
        pass

class Car(Vehicle):
    def start_engine(self):
        print("Starting the car engine")


class Bike(Vehicle):
    def start_engine(self):
        print("Starting the bike engine")
```

```python
car = Car()
car.drive()
car.start_engine()

bike = Bike()
bike.drive()
bike.start_engine()
```
