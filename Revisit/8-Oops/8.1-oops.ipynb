```python
class Dog:
    def __init__(self,name,age):
        self.name = name
        self.age = age

    def bark(self):
        print(f"Hello, I am {self.name} and my age is {self.age}")
```

```python
dog = Dog("Buddy",3)
dog.bark()
dog2 = Dog("Max",5)
dog2.bark()
```

```python
## Modelling a bank account

#Define a class for bank account
class BankAccount:
    def __init__(self,account_number,balance=0):
        self.account_number = account_number
        self.balance = balance

    def deposit(self,amount):
        self.balance += amount
        print(f"Deposited {amount}. New balance is {self.balance}")
    def withdraw(self,amount):
        if amount > self.balance:
            print(f"Insufficient balance as you are trying to withdraw {amount} but have only {self.balance}")
        else:
            self.balance -= amount
            print(f"Withdrew {amount}. New balance is {self.balance}")
    def get_balance(self):
        print(f"Current balance is {self.balance}")
```

```python
bank_account = BankAccount("123456", 1000)
bank_account.get_balance()
bank_account.deposit(500)
bank_account.get_balance()

bank_account.withdraw(20000)
bank_account.get_balance()
```
