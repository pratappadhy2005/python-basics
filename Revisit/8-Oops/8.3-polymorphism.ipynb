```python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print(f"{self.name} is speaking")
```

```python
class Dog(Animal):
    def __init__(self, name, age):
        super().__init__(name)
        self.age = age

    def speak(self):
        print(f"{self.name} is barking")
```

```python
dog = Dog("Buddy",3)
dog.speak()
```

```python
class Cat(Animal):
    def __init__(self, name, age):
        super().__init__(name)
        self.age = age
    def speak(self):
        print(f"{self.name} is meowing")
```

```python
cat = Cat("Whiskers",2)
cat.speak()
```

```python
## Function that demonstrates polymorphism
def animal_speak(animal):
    animal.speak()
```

```python
dog = Dog("Buddy",3)
animal_speak(dog)
cat = Cat("Whiskers",2)
animal_speak(cat)
```

```python
## Polymorphism with fucntions and methods
class Shape:
    def area(self):
        return "Area of the shape"

class Rectangle(Shape):
    def __init__(self,length,width):
        self.length = length
        self.width = width
    def area(self):
        return self.length * self.width

rect = Rectangle(4,5)
rect.area()
```

```python
## Polymorphism with Abstract Base Classes
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start(self):
        pass
    @abstractmethod
    def stop(self):
        pass

class Car(Vehicle):
    def start(self):
        print("Car started")
    def stop(self):
        print("Car stopped")

car = Car()
car.start()
car.stop()
```

```python
class MotorCycle(Vehicle):
    def start(self):
        print("Motorcycle started")
    def stop(self):
        print("Motorcycle stopped")
        
motorcycle = MotorCycle()
motorcycle.start()
motorcycle.stop()
```
