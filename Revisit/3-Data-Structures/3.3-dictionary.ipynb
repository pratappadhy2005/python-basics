```python
## Creating a dictionary
empty_dict = {}
type(empty_dict)
```

```python
empty_dict =dict()
type(empty_dict)
```

```python
student = {'name':'Pratap','age':25,'city':'Delhi'}
print(student)
print(type(student))
```

```python
## Accessing Dictionary Elements
student = {'name':'Pratap','age':25,'city':'Delhi'}
print(student['name'])
print(student['age'])
print(student['city'])

print("===============")
## Using get() method
print(student.get('name'))
print(student.get('age'))
print(student.get('city'))
```

```python
## Modifying Dictionary Elements
student = {'name':'Pratap','age':25,'city':'Delhi'}
print(student)
student['age'] = 26
print(student)
```

```python
## Dictionary delete, update and add
student = {'name':'Pratap','age':25,'city':'Delhi'}
print(student)
student['age'] = 26
print(student)
student['email'] = 'pratap@example.com'
print(student)

## Delete
del student['email']
print(student)
```

```python
## Common dictionary methods
keys = student.keys()
print(keys)

values = student.values()
print(values)
```

```python
## Shallow copy
student_copy = student.copy()
print(student_copy)

student_copy['age'] = 27
print(student_copy)
print(student)
```

```python
## Iterating over dictionaries
## you can use loops to iterate over dictionaries
## you can use the keys() method to get a list of keys
## you can use the values() method to get a list of values
## you can use the items() method to get a list of key-value pairs
## Iterating over keys
for keys in student_copy.keys():
    print(keys)

for values in student_copy.values():
    print(values)
```

```python
## Iterate over Key-Value Pairs
for key, value in student_copy.items():
    print(key, value)
```

```python
## Nesting Dictionaries
students = {
    "Pratap": {
        "age": 27,
        "city": "Delhi"
    },
    "Ananya": {
        "age": 26,
        "city": "Mumbai"
    }
}

for student in students:
    print(student)
    print(students[student])
    print(students[student]["age"])
```

```python
## Dictionary comprehension
sqauares = {x: x**2 for x in range(1, 6)}
print(sqauares)
```

```python
## Conditional dictionary comprehension
sqauares = {x: x**2 for x in range(1, 6) if x % 2 == 0}
print(sqauares)
```
