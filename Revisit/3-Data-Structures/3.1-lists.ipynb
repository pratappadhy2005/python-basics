```python
## Accessing List Elements
fruits = ["apple", "banana", "orange"]
print(fruits[0])
print(fruits[1])
print(fruits[2])
print(fruits[-1])
print(fruits[0:2])
```

```python
## Modify List Elements
fruits = ["apple", "banana", "orange"]
print(fruits)
fruits[0] = "mango"
print(fruits)
```

```python
## List Methods
fruits = ["apple", "banana", "orange"]
print(fruits)
fruits.append("mango")
print(fruits)
fruits.remove("banana")
print(fruits)
fruits.index("orange")
print(fruits.index("orange"))
fruits.insert(1, "peach")
print(fruits)
fruits.reverse()
print(fruits)
```

```python
## Remove and return the last item
print(fruits.pop())
print(fruits)
print(fruits.count("mango"))
fruits.sort()
print(fruits)
```

```python
## Slicing
fruits = [1,2,3,4,5,6,7,78,90]
print(fruits[0:2])
print(fruits[2:4])
print(fruits[4:])
print(fruits[:-1])
print(fruits[::2])
```

```python
## iteration
for fruit in fruits:
    print(fruit)
```

```python
number_list = [1, 2, 3, 4, 5, 6, 7, 78, 90]
for index, number in enumerate(number_list):
    print(index, number)
```

```python
## list comprehension
lst=[]
for x in range(10):
    lst.append(x**2)
    print(x)
print(lst)
```

```python
[x**2 for x in range(10)]
```

```python
## List Comprehension with condition
lst=[x for x in range(10) if x > 4]
print(lst)
```

```python
## Nested List Comprehension
lst1=[1,2,3,4,5]
lst2=["a","b","c","d","e"]

pair = [[a,b] for a in lst1 for b in lst2]
print(pair)
```

```python
## List comprehension with else
lst1=[1,2,3,4,5]
lst2=["a","b","c","d","e"]

pair = [[a,b] for a in lst1 for b in lst2 if a!=b else [a,a]]
print(pair)

```
