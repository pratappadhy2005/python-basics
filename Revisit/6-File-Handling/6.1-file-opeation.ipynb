```python
### Reading a whole file
with open('file.txt', 'r') as file_object:
    contents = file_object.read()
    print(contents)
```

```python
## Read line by line
with open('file.txt', 'r') as file_object:
    for line in file_object:
        print(line.strip())
```

```python
## Writing a file (Overwriting)
with open('file.txt', 'w') as file_object:
    file_object.write("Hello World!\n")
    file_object.write("This is a new line!\n")
```

```python
with open('file.txt', 'r') as file_object:
    for line in file_object:
        print(line.strip())
```

```python
## Appending to a file
with open('file.txt', 'a') as file_object:
    file_object.write("This is a new line!\n")
```

```python
with open('file.txt', 'r') as file_object:
    for line in file_object:
        print(line.strip())
```

```python
## Write list of lines to a file
with open('file.txt', 'w') as file_object:
    file_object.writelines(["Hello World!\n", "This is a new line!\n"])
```

```python
with open ('file.txt', 'r') as file_object:
    for line in file_object:
        print(line.strip())
```

```python
## Binary files
with open('example.bin', 'wb') as f:
    f.write(b'Hello, world!')
```

```python
with open('example.bin', 'rb') as f:
    data = f.read()
    print(data)
    print(data.decode('utf-8'))
```

```python
## Read the content from a source text file and write in another file
with open('source.txt', 'r') as f:
    content = f.read()

with open('destination.txt', 'w') as f:
    f.write(content)

with open('destination.txt', 'r') as f:
    content = f.read()
    print(content)
```

```python
## Writing and then reading from a file
with open('temp.txt', 'w+') as file:
    file.write('Hello, World!')
    file.seek(0)
    content = file.read()
    print(content)
```

```python
## Create a new directory
import os

os.mkdir('new_dir')

## create a new file in the new directory
with open(os.path.join('new_dir', 'temp.txt'), 'w+') as file:
    file.write('Hello, World!')
```

```python
## list all files in the new directory
import os

os.listdir('new_dir')
```
