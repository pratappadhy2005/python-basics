```python
def gen_values():
    yield 1
    yield 2
    yield 3

gen_value_object = gen_values()
# print(next(gen_value_object))
# print(next(gen_value_object))
# print(next(gen_value_object))
```

```python
for value in gen_value_object:
    print(value)
```

```python
def get_primes_list(start, end):
    for num in range(start, end+1):
        if num > 1:
            for i in range(2, num):
                if (num % i) == 0:
                    break
            else:
                yield num


get_primes = get_primes_list(1, 100)
for prime in get_primes:
    print(prime)
```
