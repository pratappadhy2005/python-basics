```python
## For loop
for i in range(10):
    print(i)
```

```python
for i in range(0,6):
    print(i)
```

```python
for i in range(1,10,2):
    print(i)
```

```python
for i in range(10,1,-1):
    print(i)
```

```python
## Strings
str="Hello World"
print(str)
for i in str:
    print(i)
```

```python
## while loop
i=0
while i<10:
    print(i)
    i=i+1
```

```python
count=0
while count%2==0:
    print(count)
    count=count+1
```

```python
## Loop control statements

## Break exits the loop prematurely

for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
## Continue skips the current iteration
for i in range(10):
    if i == 5:
        continue
    print(i)
```

```python
## pass
## The pass state is a null operation; nothing happens when it is executed.
for i in range(10):
    pass
```

```python
## Nested Loops
for i in range(5):
    for j in range(5):
        print(f"{i} : {j}")
```

```python
## Examples: Calculate the sum of first N natural numbers using a while and for loop
n =10
sum=0
count=1

while count<=n:
    sum=sum+count
    count=count+1

print(f"sum={sum}")
```

```python
n =10
sum=0
count=1

for i in range(1,n+1):
    sum=sum+i
print(f"sum={sum}")
```

```python
## Display the prime number between 1 and 100
for i in range(1,101):
    if i%2!=0:
        print(i)
```
