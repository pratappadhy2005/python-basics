```python
## Exception Handling try, except block
try:
    a=b
except NameError:
    print("Variable b is not defined")
```

```python
try:
    a=b
except NameError as ex:
    print(ex)
```

```python
try:
    #result =1/0
    a=b
except ZeroDivisionError as ex:
    print(ex)
    print("Please check the divisor")
except NameError as ex:
    print(ex)
    print("Please check the variable")
except Exception as ex:
    print(ex)
    print("Please check the code")
```

```python
try:
    num = int(input("Enter a number: "))
    result = 8/num
    print(result)
except ValueError:
    print("Invalid input. Please enter a number.")
except ZeroDivisionError:
    print("Error: Division by zero.")
```

```python
## try, except, else
try:
    a=b
except NameError:
    print("Variable b is not defined")
else:
    print(a)
```

```python
## try, except, else and finally
try:
    a=b
except NameError:
    print("Variable b is not defined")
else:
    print(a)
finally:
    print("End of try-except-else-finally block")
```

```python
## File handling with try, except, else and finally
try:
    f=open("test1.txt","w")
    f.write("Hello World")
except:
    print("Error in writing to file")
else:
    print("File written successfully")
finally:
    f.close()
```

```python
## File read with try, except, else and finally
try:
    f=open("test1.txt","r")
    print(f.read())
except Exception as e:
    print("Error in reading file:",e)
else:
    print("File read successfully")
finally:
    f.close()
```
