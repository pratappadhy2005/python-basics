```python
def square(num):
    return num*2

def cube(num):
    return num**3

def transform(num_list, func):
    return [func(num) for num in num_list]  

transform_list = transform([1, 2, 3, 4, 5], square)
print(transform_list)


transform_list_cube = transform([1, 2, 3, 4, 5], cube)
print(transform_list_cube)


transform_list_lambda_cube = transform([1,2,3,4,5,], lambda num: num*3)
print(transform_list_lambda_cube)

transform_list_lambda_square = transform([1,2,3,4,5,], lambda num: num*2)
print(transform_list_lambda_square) 
```

```python
nums_list = [1,2,3,4,5]
map_list = map(lambda num: num**3, nums_list)
print(list(map_list))   
```

```python
filter_list = filter(lambda num: num%2==0, nums_list)
print(list(filter_list))

```
