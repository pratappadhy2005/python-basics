```python
## Why functions
num=24

if num%2==0:
    print("Even")
else:
    print("Odd")
```

```python
def isNumberIsEven(num):
    if num%2==0:
        return True
    else:
        return False

print(isNumberIsEven(10))
```

```python
## Multiple parameters
def add(num1, num2):
    return num1+num2

print(add(10, 20))
```

```python
## Default parameters
def add(num1, num2=10):
    return num1+num2
        
print(add(10, 20))
print(add(10))
```

```python
## Variable number of arguments
def add(*args):
    result = 0
    for num in args:
        result += num
    return result
        
print(add(10, 20))
print(add(10, 20, 30))
```

```python
## Keyword arguments
def kadd(**kwargs):
    result = 0
    for num in kwargs.values():
        result += num
    return result

print(kadd(a=10, b=20))
```

```python
## combine positional and keyword arguments
def combine(*args, **kwargs):
    result = 0
    for num in args:
        result += num
    for num in kwargs.values():
        result += num
    return result

print(combine(10, 20, a=30, b=40))
```

```python
## Return statement
def add(a, b):
    return a + b

print(add(10, 20))
```
