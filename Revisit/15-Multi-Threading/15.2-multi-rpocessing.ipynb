```python
import time
from concurrent.futures import ProcessPoolExecutor, as_completed

def do_cpu_work(task_id: int, duration: float = 2.0) -> str:
    result = 0
    for i in range(1000000):
        result += i
    return f"Task {task_id} completed after {duration} seconds with result {result}"
   

def run_multi_process(tasks: int=5, max_workers: int=5) -> list[str]:
    results = []
    with ProcessPoolExecutor(max_workers=max_workers) as executor:
        future_to_task_id = [executor.submit(do_cpu_work, task_id, 1000000) for task_id in range(tasks)]
        for future in as_completed(future_to_task_id):
            task_id = future_to_task_id.index(future)
            try:
                result = future.result()
                results.append(result)
            except Exception as e:
                print(f"Task {task_id} generated an exception: {e}")
    return results

if __name__ == "__main__":      
    start_time = time.time()
    results = run_multi_process(tasks=10, max_workers=2)    
    end_time = time.time()
    print(results)
    print(f"Total time taken: {end_time - start_time} seconds")
```
